# **3.** Stability of explanations: comparison of LIME decompositions

In the picture below we have LIME explanations computed two times for each of three observations. LIME explanations can be unstable due to the Generation Step: each time when LIME is runned it generates other random variables to fit the hyperplane to the model surface. In this case for the first observation the order of variables differs in two estimations but most important features are the same. For the second observation the variable 'National Team Jersey Number' contributed positively for one case and negatively for another. It was the case also for the third case. In this very case it may lead to the conclusion that the variable 'National Team Jersey Number' may be in some wide sense somehow misleading and when we consider the meaning of this feature (it is simply the number on the player's T-shirt) we can come to the conclusion that we can in fact get rid of this feature.
