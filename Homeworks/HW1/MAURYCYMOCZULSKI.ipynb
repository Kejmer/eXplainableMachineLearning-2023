```python
import pandas as pd
import plotly.express as px
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
fifa = pd.read_csv('Fifa 23 Players Data.csv')
fifa.head()
```

```python
fifa_num = fifa.select_dtypes(['number'])
```

```python
training_data = fifa_num.sample(frac=0.8, random_state=25)
testing_data = fifa_num.drop(training_data.index)
```

```python
testing_data
```

```python
X_train = training_data.drop(['Overall', 'Potential'], axis=1)
X_test = testing_data.drop(['Overall', 'Potential'], axis=1)
```

I dicieded to take all numerical values expect potential becouse I believe that this column should be explained and not used for explanations

```python
Y_train = training_data['Overall']
Y_test = testing_data['Overall']
```

I want to explain Overall by data using Linear Regresion

```python
linreg = LinearRegression().fit(X_train, Y_train)
linreg.score(X_test, Y_test)
```

Score is not bad but I belevie It could be better we should delete less important columns to improve that score

```python
linreg.coef_
```

```python
linreg.intercept_
```

```python
np.argmax(linreg.coef_)
```

```python
X_test.iloc[:,[10]]
```

As we can see the most important column is Skill Moves

```python
np.argmin(linreg.coef_)
```

```python
X_test.iloc[:,[11]]
```

And the least important column is International Reputation

Now lets try to explain Overall using random forest algorythm

```python
rfr = RandomForestRegressor(max_depth=10, random_state=0)
rfr.fit(X_train, Y_train)
```

```python
rfr.score(X_test, Y_test)
```

```python
rfr.feature_importances_
```

```python
np.argmax(rfr.feature_importances_)
```
